```python
from scipy.spatial.distance import pdist, squareform

def takens_embed(x, m=4, tau=1):
    x = np.asarray(x)
    N = len(x) - (m-1)*tau
    if N <= 2:
        raise ValueError("Time series too short for embedding with m={}, tau={}".format(m, tau))
    return np.column_stack([x[i:i+N] for i in range(0, m*tau, tau)])

def correlation_sum(X, r_vals, theiler=0):
    # pairwise distances
    D = squareform(pdist(X, metric='euclidean'))
    N = len(X)

    # build upper-triangular mask and drop Theiler window |i-j| <= theiler
    mask = np.triu(np.ones_like(D, dtype=bool), 1)
    if theiler > 0:
        for k in range(-theiler, theiler+1):
            if k == 0: 
                continue
            i = np.arange(N)
            j = i + k
            valid = (j >= 0) & (j < N)
            mask[i[valid], j[valid]] = False

    dists = D[mask]
    C = np.array([(dists < r).mean() for r in r_vals], dtype=float)
    return C

def corr_dim_gp(x, m=4, tau=1, theiler=0, n_r=40,
                r_frac_range=(0.01, 0.3), fit_slice=(12, 28)):
    """
    x: 1D time series
    m,tau: embedding params (your request: m=4, tau=1)
    theiler: samples to exclude around the diagonal (set ~ one beat in samples if applicable)
    n_r: # of radii
    r_frac_range: radii span as fractions of attractor diameter
    fit_slice: index range on log-log curve used for linear fit (scaling region)

    returns: D2 (slope), (r_vals, C_vals), fit_indices
    """
    X = takens_embed(x, m=m, tau=tau)
    # choose radii relative to attractor diameter
    D = squareform(pdist(X))
    diam = D.max()
    r_vals = np.logspace(np.log10(diam*r_frac_range[0]),
                         np.log10(diam*r_frac_range[1]),
                         n_r)
    C = correlation_sum(X, r_vals, theiler=theiler)

    # fit in the chosen scaling region
    i1, i2 = fit_slice
    i1 = max(0, i1); i2 = min(len(r_vals), i2)
    lx = np.log(r_vals[i1:i2])
    ly = np.log(C[i1:i2] + 1e-15)
    slope, intercept = np.polyfit(lx, ly, 1)

    return slope, (r_vals, C), (i1, i2), intercept


```
