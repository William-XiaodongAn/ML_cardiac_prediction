```python

data = []
end_of_first_dt = False
path = r"..\data\tau_d_0.4095_temporal_APD_step_4.csv"
#path = r"F://result/tau_pw_350_tau_d_0.406_spatial_20e4_32e4_8_8_APD_step_4.csv"

number_of_points = 0
end_of_first_dt = False

data = []
with open(path) as csv_file: # totally 501 points
    csv_reader = csv.reader(csv_file, delimiter=',')
    
    for row in csv_reader:
        for i in range(len(row)):
            for j in row[i]:
                if j == ';':
                    row[i] = row[i].replace(';', '')
                    if end_of_first_dt == False:
                        end_of_first_dt = True
                        number_of_points = i
                    
            if row[i] != '' and i%4 != 0: 
                data.append(float(row[i]))    
max_value = max(data)
min_value = min(data)
# if we input whole texture
if (number_of_points * 3 / 4).is_integer():
    number_of_points = int( number_of_points * 3 / 4)
else:
    print('error! non-integar number of points')
# if we input whole texture

if (len(data)/number_of_points).is_integer():
    rows, cols=number_of_points,int(len(data)/number_of_points)
else:
    print('error! non-integar time steps',len(data),number_of_points) 
    
voltage = np.zeros([rows,cols])

for i in range(rows):
    for j in range(cols):

        voltage[i][j] = data[j * rows +  i]

print(number_of_points)
```

```python


path = "output.dat.lyap"   # <-- change if your file is elsewhere

dim_to_pick = 4
vals = []
idxs = []

with open(path, "r") as f:
    lines = f.readlines()

# Find the first header line that matches "dim= 4"
header_re = re.compile(r"#epsilon=\s*([^\s]+)\s+dim=\s*(\d+)")
in_block = False

for i, line in enumerate(lines):
    m = header_re.match(line.strip())
    if m:
        # Enter a new block when we see a header
        eps = m.group(1)
        dim = int(m.group(2))
        if not in_block and dim == dim_to_pick:
            # Start collecting from the next lines
            in_block = True
            start_i = i + 1
            continue
        elif in_block:
            # We already collected the first dim=4 block; a new header means stop
            break

    # If we're inside the first dim=4 block, collect numeric rows until blank/header
    if in_block:
        s = line.strip()
        if not s or s.startswith("#"):
            break
        parts = s.split()
        if len(parts) >= 3:
            try:
                idx = int(parts[0])
                mid = float(parts[1])  # the "middle value" between idx and 2446
                # tail = int(parts[2]) # typically 2446; not used for plotting here
                idxs.append(idx)
                vals.append(mid)
            except ValueError:
                # hit a non-data line inside the block; stop
                break

if not vals:
    raise RuntimeError("Did not find a 'dim= 4' block or it contained no data.")

# Plot: middle column vs idx
plt.figure()
plt.plot(idxs, vals)
plt.xlabel("idx (evolution step r)")
plt.ylabel("mean log distance (middle column)")
plt.title("First dim=4 block (epsilon = {})".format(eps))
plt.tight_layout()
plt.show()
```

```python
def lyap_from_segment(idx, val, r1, r2, dt=0.004):
    """
    idx : 1D array of evolution steps (r)
    val : 1D array of <ln distance> at those steps
    r1, r2 : inclusive window bounds in idx over which the curve is ~linear
    dt : sampling interval in seconds (default 4 ms)
    Returns: dict with slope_per_step, lambda_per_s, T_lyap_s, intercept, R2
    """
    idx = np.asarray(idx); val = np.asarray(val)
    m = (idx >= r1) & (idx <= r2)
    x = idx[m].astype(float)
    y = val[m].astype(float)

    # Ordinary least squares via polyfit
    b, a = np.polyfit(x, y, 1)   # y  a + b*x
    yhat = a + b*x
    ss_res = np.sum((y - yhat)**2)
    ss_tot = np.sum((y - y.mean())**2)
    R2 = 1.0 - ss_res/ss_tot if ss_tot > 0 else np.nan

    slope_per_step = b                     # nats per step
    lam_per_s = slope_per_step / dt        # s^-1
    T_lyap_s = 1.0 / lam_per_s             # seconds

    return {
        "slope_per_step": slope_per_step,
        "lambda_per_s": lam_per_s,
        "T_lyap_s": T_lyap_s,
        "intercept": a,
        "R2": R2,
        "n_points": int(m.sum())
    }

res = lyap_from_segment(idxs, vals, r1=0, r2=50, dt=0.004)
print(res)
```
